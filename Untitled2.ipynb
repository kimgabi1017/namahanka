```python
import pandas as pd
from transformers import MarianMTModel, MarianTokenizer

def translate_text(text, model, src_lang, tgt_lang):
    # 소스 언어 및 대상 언어 설정
    src_text = f">>{src_lang}<< {text}"
    src_ids = model.encode(src_text, return_tensors="pt")

    # 번역 수행
    with model.as_target_tokenizer():
        translation_ids = model.generate(src_ids, max_length=50, num_beams=5, length_penalty=0.6)

    # 번역 결과 디코딩
    translation = model.decode(translation_ids[0], skip_special_tokens=True)
    return translation

def translate_csv(input_csv, output_csv, model_name, src_lang, tgt_lang):
    # CSV 파일 읽기 (탭으로 구분)
    df = pd.read_csv(input_csv, delimiter='\t')

    # 모델과 토크나이저 로드
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    # 각 행의 일본어 문장을 한국어로 번역
    translations = []
    for index, row in df.iterrows():
        japanese_text = row['ja']
        translation = translate_text(japanese_text, model, src_lang, tgt_lang)
        translations.append(translation)

    # 번역 결과를 DataFrame에 추가
    df[tgt_lang] = translations

    # 번역 결과를 CSV 파일로 저장
    df.to_csv(output_csv, index=False, sep='\t')

# 입력 및 출력 파일 경로 설정
input_csv = 'your_input_file.csv'  # 입력 CSV 파일 경로
output_csv = 'your_output_file.csv'  # 출력 CSV 파일 경로

# 모델 및 언어 설정
model_name = "Helsinki-NLP/opus-mt-ja-ko"  # 일본어에서 한국어로 번역하는 모델 사용
src_lang = 'ja'  # 일본어 코드
tgt_lang = 'ko'  # 한국어 코드

# 번역 수행 및 결과 저장
translate_csv(input_csv, output_csv, model_name, src_lang, tgt_lang)
```
